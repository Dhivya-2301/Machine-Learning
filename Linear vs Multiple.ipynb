```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("study_hours.csv")
x=ds.iloc[:,0:1].values
y=ds.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x,y)
y_lin=l.predict(x)

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)

poly_reg=LinearRegression()
poly_reg.fit(x_poly,y)
y_poly=poly_reg.predict(x_poly)
```

```python
plt.scatter(x,y)
plt.plot(x,y_lin)
plt.plot(x,y_poly)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear vs Polynomial regression")
plt.show()
```

```python
from sklearn.metrics import r2_score
print("Linear R2 Score:", r2_score(y, y_lin))
print("Polynomial R2 Score:", r2_score(y, y_poly))
```

```python
import joblib

joblib.dump(poly, "poly_transform.pkl")
joblib.dump(poly_reg, "polynomial_regression.pkl")

print("Successfully saved")
```
